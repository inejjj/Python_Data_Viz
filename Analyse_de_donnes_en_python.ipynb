```python
#4) Ecrivez une fonction qui prend en entrée un pays et affiche:
#Si elle a plus que 4 licornes compris: "Le pays X a X licornes pour une valorisation totale de X $B"
#Si elle a entre 1 et 3 licornes, donner les détails: "Le pays X n'a que X licornes: ?, ? et ?"
#Si elle n'en a aucune: "Le pays X n'a aucune licorne"

def compteur_licornes(pays):
  #Nous allons définir une variable pour bien filtrer sur le pays de notre paramètre
  filtre_pays = df_unicorns[df_unicorns['Country'] == pays]

  #Nous allons compter le nombre de pays dans notre sélection
  total_licornes = len(filtre_pays)

  #Nous allons sommer dès maintenant la valorisation
  total_valorisation = filtre_pays['Valuation ($B)'].sum()

  #Maintenant nous pouvons préparer nos sorties en fonction du nombre de licornes
  if total_licornes > 4:
    print("Le pays", pays, "a", total_licornes, "licornes pour une valorisation totale de", total_valorisation, "$B")

  elif total_licornes >= 1:
  #Nous devons créer une nouvelle variable permettant de citer le nom des licornes qu'il y en ait 1, 2 ou 3
    licornes = ', '.join(filtre_pays['Company'])
    print("Le pays", pays, "n'a que", total_licornes, "licorne(s) :", licornes)

  else:
    print("Le pays", pays, "n'a aucune licorne")

compteur_licornes("China")
compteur_licornes("Croatia")
compteur_licornes("Malaysia")
```

```python
#5) Ecrivez une fonction qui prend en entrée un secteur d'activité et qui renvoie le pays avec le plus de licornes

def max_licornes(secteur):
  #Nous allons d'abord nous assurer que l'user renseigne un paramètre présent dans le champ Industry
  secteurs_valides = ['Artificial intelligence', 'Other',
       'E-commerce & direct-to-consumer', 'Fintech',
       'Internet software & services',
       'Supply chain, logistics, & delivery',
       'Data management & analytics', 'Edtech', 'Hardware',
       'Consumer & retail', 'Health', 'Auto & transportation',
       'Cybersecurity', 'Mobile & telecommunications', 'Travel',
       'Internet Software Services']

  if secteur not in secteurs_valides:
        print("Le secteur n'est pas valide. Veuillez renseigner l'un des secteurs suivants:", ', '.join(secteurs_valides))
        return None

  #Maintenant, filtrons sur le secteur renseigné
  licornes_secteur = df_unicorns[df_unicorns['Industry'] == secteur]

  #Pour être sûr de sélectionner le bon pays, nous retournons avec l'index du pays qui a la valeur la plus élevée
  pays_max_licornes = licornes_secteur['Country'].value_counts().idxmax()
  max_licornes = licornes_secteur['Country'].value_counts().max()

  print("Le pays", pays_max_licornes, "a le plus de licornes dans le secteur", secteur, "avec" , max_licornes, "licornes")

max_licornes('Edtech')
max_licornes('Hardware')
max_licornes('Data management and analytics')
max_licornes('Data management & analytics')
```

## **Analyse géographique**

```python
tournages_par_quartier = df_tournage['Localisation de la scène'].value_counts()
top_quartiers = tournages_par_quartier.head(10)

plt.figure(figsize=(12, 6))
top_quartiers.sort_values().plot(kind='barh')
plt.title('Top 10 des lieux les plus populaires pour les tournages')
plt.xlabel('Nombre de tournages')
plt.ylabel('Lieux')
plt.grid(axis='x')
plt.show()
```

```python
#Quels sont les arrondissements les plus représentés ?

tournages_par_arrondissement = df_tournage['Code postal'].value_counts()

plt.figure(figsize=(12, 6))
tournages_par_arrondissement.sort_values().plot(kind='barh')
plt.title('Classement des arrondissements les plus représentés')
plt.xlabel('Nombre de tournages')
plt.ylabel('Arrondissement')
plt.grid(axis='x')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df_tournage['Type de tournage'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Répartition des types de tournages')
plt.show()
```

```python
#En faisant nos recherches pour détailler le graph précédent, nous avons découvert la fonction crosstab de pandas:
#https://www.geeksforgeeks.org/using-pandas-crosstab-to-create-a-bar-plot/

plt.figure(figsize=(12, 8))
pd.crosstab(df_tournage['Année du tournage'], df_tournage['Type de tournage']).plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Répartition des types de tournages au fil des ans')
plt.xlabel('Année du tournage')
plt.ylabel('Nombre de tournages')
plt.legend(title='Type de tournage')
plt.show()
```

```python
# Evolution du type de tournage en fonction des années
tournages_par_annee = df_tournage.groupby(['Année du tournage', 'Type de tournage']).size().unstack(fill_value=0)

# Créer le graphique en courbe
tournages_par_annee.plot(kind='line', marker='o')
plt.title('Évolution du type de tournage à Paris par année')
plt.xlabel('Année')
plt.ylabel('Nombre de tournages')
plt.legend(title='Type de tournage')
plt.show()
```

*   La vue en barre du graph précédent est transformée en courbe avec les détails par type de tournage qui nous donne des explications de l'évolution du nombre de tournage à Paris

# Analyse détailléé par arrondissement

```python
def lieux_plus_utilises_par_code_postal(code_postal):
    # on filtre le dataframe pour le code postal spécifié
    filtre_code_postal = df_tournage[df_tournage['Code postal'] == code_postal]

    # on sélectionne les trois lieux les plus utilisés
    adresses = ', '.join(filtre_code_postal['Localisation de la scène'].head(3))

    print("Dans le", code_postal, "les lieux les plus utilisés sont :", adresses)

lieux_plus_utilises_par_code_postal(75020)
lieux_plus_utilises_par_code_postal(75014)
```

```python
# Fonction qui prend en entrée l’année parmi les années dans la base et rend le top 5 arrondissements les plus tournés


def top_5_arrondissements_par_annee_graph_stacked_vertical(annee):
    # Filtrer le DataFrame pour obtenir les données de l'année spécifiée
    df_annee = df_tournage[df_tournage['Année du tournage'] == annee]

    # Regrouper par arrondissement et type de tournage, compter le nombre de films par arrondissement
    films_par_arrondissement = df_annee.groupby(['Code postal', 'Type de tournage']).size().unstack(fill_value=0)

    # Obtenir les trois premiers arrondissements avec le plus grand nombre de films tournés
    top_5_arrondissements = films_par_arrondissement.sum(axis=1).nlargest(5).index

    # Filtrer le DataFrame pour ne conserver que les trois premiers arrondissements
    top_5_df = films_par_arrondissement.loc[top_5_arrondissements]

    # Trier les arrondissements par nombre total de films tournés (ordre décroissant)
    top_5_df = top_5_df.loc[top_5_df.sum(axis=1).sort_values(ascending=False).index]

    # Créer un graphique en barres verticales empilées pour le top 3 des arrondissements
    top_5_df.plot(kind='bar', stacked=True)
    plt.title("Top 5 des arrondissements en " +str(annee) + " par nombre de films tournés et type de tournage")
    plt.xlabel('Arrondissement')
    plt.ylabel('Nombre de films')
    plt.legend(title='Type de tournage')

    plt.show()


# Appel de la fonction avec l'année 2019
top_5_arrondissements_par_annee_graph_stacked_vertical(2019)
```

*   Nous avons cette fontion qui nous permet de voir en détail du Top 5 arrondissments par année après avoir constaté la globale

```python
# Fonction qui prend en entrée un arrondissement et qui rend un grap de courbe de l'évolution du type de tournage en fonction des années


def evolution_type_tournage_par_arrondissement(arrondissement):
    # Filtrer le DataFrame pour obtenir les données de l'arrondissement spécifié
    df_arrondissement = df_tournage[df_tournage['Code postal'] == arrondissement]

    # Regrouper par année et type de tournage, compter le nombre de films par année
    tournages_par_annee = df_arrondissement.groupby(['Année du tournage', 'Type de tournage']).size().unstack(fill_value=0)

    # Créer un graphique de courbe pour l'évolution du type de tournage en fonction des années
    tournages_par_annee.plot(marker='o', linestyle='-')
    plt.title("Évolution du type de tournage à Paris, arrondissement " + str(arrondissement))
    plt.xlabel('Année')
    plt.ylabel('Nombre de films')
    plt.legend(title='Type de tournage')
    plt.grid(True)
    plt.show()

# Appel de la fonction avec l'arrondissement 75001
evolution_type_tournage_par_arrondissement(75001)
```
